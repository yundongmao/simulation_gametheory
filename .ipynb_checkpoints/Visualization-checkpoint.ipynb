```python
import json
import numpy as np
import matplotlib.pyplot as plt
import collections
from mpl_toolkits.mplot3d import Axes3D
f = open("simulation_0001")
lines = f.readlines()
f.close()
point = collections.defaultdict(dict)
for line in lines:
    m = json.loads(line)
    point[m["p"]][m["reward"]] = m["fixation_prob"]

X = sorted(point.keys())
Y = sorted(point[X[0]].keys())
Z = [[] for i in range(len(Y))]
for x in point:
    for i,y in enumerate( sorted(point[x].keys())):
        Z[i].append(point[x][y])
X = np.array(X)
Y = np.array(Y)
Z = np.array(Z)
X, Y = np.meshgrid(X, Y)
# print X
# print Y
# print Z.shape
# print X.shape
# print Y.shape


from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='rainbow')

plt.show()

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='rainbow')
plt.show()
```

```python
import json
import numpy as np
import matplotlib.pyplot as plt
import collections
from mpl_toolkits.mplot3d import Axes3D
f = open("simulation_BaraAlbert")
lines = f.readlines()
f.close()
point = collections.defaultdict(dict)
for line in lines:
    m = json.loads(line)
    point[m["outDegree"]][m["size"]] = m["fixation_prob"]

X = sorted(point.keys())
x_v_i = {}
for i,v in enumerate(X):
    x_v_i[v] = i;
Y = sorted(point[X[0]].keys())
y_v_i = {}
for i,v in enumerate(Y):
    y_v_i[v] = i;

Z = [[] for i in range(len(Y))]
for x in point:
    temp = sorted(point[x].keys())
    for i in range(0,y_v_i[temp[0]]):
        Z[i].append(0.0)
    for y in temp:
        Z[y_v_i[y]].append(point[x][y])
    
        
X = np.array(X)
Y = np.array(Y)
Z = np.array(Z)
X, Y = np.meshgrid(X, Y)

print Z.shape
print X.shape
print Y.shape


from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='rainbow')

plt.show()
```

```python


import numpy as np
import matplotlib.pyplot as plt
import json
f = open("simulation_week7_Erdo_003")
lines = f.readlines()
# print lines
x = []
y = []
z = []



for line in lines:
    m = json.loads(line)
    y.append(m["fixation_prob"])
    x.append(m["size"])
    r = m["reward"]
    N = m["size"]
    z.append((1-1/(float(r)))/((1-1/(float(r)**(N)))))

plt.plot(x, y,label = "fixation probability on graph")
plt.plot(x, z,label = "fixation probability on well-mixed population")
plt.show()
```
