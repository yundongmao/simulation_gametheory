```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-4, 4, 1)
Y = np.arange(-4, 4, 2)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.cos(R)
print X
print Y
print Z

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='rainbow')

plt.show()
```

```python
import json
import collections
f = open("simulation_0001")
lines = f.readlines()
f.close()
point = collections.defaultdict(dict)
for line in lines:
    m = json.loads(line)
    point[m["p"]][m["reward"]] = m["fixation_prob"]

X = sorted(point.keys())
Y = sorted(point[X[0]].keys())
Z = [[] for i in range(len(Y))]
for x in point:
    for i,y in enumerate( sorted(point[x].keys())):
        Z[i].append(point[x][y])
X = np.array(X)
Y = np.array(Y)
Z = np.array(Z)
X, Y = np.meshgrid(X, Y)
print X
print Y
print Z.shape
print X.shape
print Y.shape


from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='rainbow')

plt.show()

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='rainbow')
plt.show()
```
